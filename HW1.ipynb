```python
import pandas as pd
import numpy as np
import sklearn.feature_selection as sk1
import sklearn.linear_model as sk2
import sklearn.metrics as sk3
```

```python
data = pd.read_csv('pima-indians-diabetes.data.csv')
```

```python
print(data)
print(data.shape)
```

```python
data.head()
```

```python
data2 = data.to_numpy()
```

```python
data2
```

```python
datay = [i for i in data['Outcome']]
```

```python
datay
```

```python
datax = np.delete(data2,8,1)
```

```python
select = sk1.SelectKBest(score_func = sk1.f_classif, k = 3)
data3 = select.fit_transform(datax,datay)
```

```python
data3.shape
```

```python
datax[0][select.get_support()]
```

```python
select = sk1.SelectKBest(score_func = sk1.f_classif, k = 4)
data4 = select.fit_transform(datax,datay)
```

```python
data4.shape
```

```python
datax[0][select.get_support()]
```

```python
model = sk2.LogisticRegression()

x = []

for i in data2:
    x.append([i[0],i[5]])
    
model.fit(x,datay)


ypred = model.predict(x)

print("precision: ", sk3.precision_score(datay,ypred),", recall: ", sk3.recall_score(datay,ypred), "F1 score: ", sk3.f1_score(datay,ypred) )
```
