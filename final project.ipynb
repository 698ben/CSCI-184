```python
temp = []
datax = []
datay = []
comp = True


for j in range(0,48153):
    for i in data.columns:
        if(i != 'Id' and i != 'date' and i != 'razon_removed' and i != 'time' and i != 'age_cat' and i != 'underline_disease' and i != 'Transplant' and i != 'subregion' and i != 'cPRA_cat' and i != 'DR_00' and i != 'B_00' and i != 'A_00'):
            if(i == 'race'):
                if (data[i][j] == 'Branca'):
                    temp.append(0)
                elif(data[i][j] == 'Parda'):
                    temp.append(1)
                elif(data[i][j] == 'Negra'):
                        temp.append(2)
                else:
                        temp.append("nan")
            elif(i == 'sex'):
                if(data[i][j] == 'M'):
                    temp.append(0)
                elif(data[i][j] == 'F'):
                    temp.append(1)
            elif(i == 'Blood_type'):
                if(data[i][j] == 'A'):
                    temp.append(0)
                elif(data[i][j] == 'O'):
                    temp.append(1)
                elif(data[i][j] == 'B'):
                    temp.append(2)
                else:
                    temp.append("nan")
            elif(i == 'Transplant_Y_N'):
                if(data[i][j] == 'Nao'):
                    temp.append(0)
                elif(data[i][j] == 'Sim'):
                    temp.append(1)
                else:
                    temp.append("nan")
            elif(i == 'removed_list'):
                if(data[i][j] == 'Nao'):
                    temp.append(0)
                elif(data[i][j] == 'Sim'):
                    temp.append(1)
                else:
                    temp.append("nan")
            elif(i == 'priorization'):
                if(data[i][j] == 'Nao'):
                    temp.append(0)
                elif(data[i][j] == 'Sim'):
                    temp.append(1)
                else:
                    temp.append("nan")
            elif(i == 'gestation'):
                if(data[i][j] == 'Nao'):
                    temp.append(0)
                elif(data[i][j] == 'Sim'):
                    temp.append(1)
                else:
                    temp.append("nan")
            elif(i == 'prior_transplant'):
                if(data[i][j] == 'Nao'):
                    temp.append(0)
                elif(data[i][j] == 'Sim'):
                    temp.append(1)
                else:
                    temp.append("nan")
            elif(i == 'chagas'):
                if(data[i][j] == 'Nao'):
                    temp.append(0)
                elif(data[i][j] == 'Sim'):
                    temp.append(1)
                else:
                    temp.append("nan")
            elif(i == 'patient_still_on_list'):
                if(data[i][j] == 'Nao'):
                    temp.append(0)
                elif(data[i][j] == 'Sim'):
                    temp.append(1)
                else:
                    temp.append("nan")
            elif(i == 'death'):
                if(data[i][j] == 'Nao'):
                    temp.append(0)
                elif(data[i][j] == 'Sim'):
                    temp.append(1)
                else:
                    temp.append("nan")
            elif(i == 'anti.HBc'):
                if(data[i][j] == 'Nao'):
                    temp.append(0)
                elif(data[i][j] == 'Sim'):
                    temp.append(1)
                else:
                    temp.append("nan")
            elif(i == 'anti.HCV'):
                if(data[i][j] == 'Nao'):
                    temp.append(0)
                elif(data[i][j] == 'Sim'):
                    temp.append(1)
                else:
                    temp.append("nan")
            elif(i == 'agHBs'):
                if(data[i][j] == 'Nao'):
                    temp.append(0)
                elif(data[i][j] == 'Sim'):
                    temp.append(1)
                else:
                    temp.append("nan")
            else:
                temp.append(data[i][j])
        if(i == 'time'):
            datay.append([data[i][j]])
    if(comp):
        datax.append(temp)
    temp = []
    comp = True


```

```python
def concat(arr,arr2):
    arr3 = []
    for x in arr:
        arr3.append(x)    
    for x in arr2:
        arr3.append(x)
    print(type(arr))
    return arr3
```

```python
for i in range(len(datax) - 1):
    for j in range(len(datax[i]) - 1):
        if (str(datax[i][j]) == 'nan'):
            datax = concat(datax[::i + 1],datax[i + 1::])
            
datax = datax[0::1]
```
