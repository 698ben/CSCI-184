```python
temp = []
datax = []
datay = []
comp = True


for j in range(0,43337):
    for i in train.columns:
        if(i != 'Id' and i != 'date' and i != 'razon_removed' and i != 'time' and i != 'Time_Tx' and i != 'age_cat' and i != 'underline_disease' and i != 'Transplant' and i != 'subregion' and i != 'cPRA_cat' and i != 'DR_00' and i != 'B_00' and i != 'A_00'):
            if(i == 'race'):
                if (train[i].array[j] == 'Branca'):
                    temp.append(0)
                elif(train[i].array[j] == 'Parda'):
                    temp.append(1)
                elif(train[i].array[j] == 'Negra'):
                        temp.append(2)
                else:
                        temp.append("nan")
            elif(i == 'sex'):
                if(train[i].array[j] == 'M'):
                    temp.append(0)
                elif(train[i].array[j] == 'F'):
                    temp.append(1)
            elif(i == 'Blood_type'):
                if(train[i].array[j] == 'A'):
                    temp.append(0)
                elif(train[i].array[j] == 'O'):
                    temp.append(1)
                elif(train[i].array[j] == 'B'):
                    temp.append(2)
                else:
                    temp.append("nan")
            elif(i == 'Transplant_Y_N'):
                if(train[i].array[j] == 'Nao'):
                    temp.append(0)
                elif(train[i].array[j] == 'Sim'):
                    temp.append(1)
                else:
                    temp.append("nan")
            elif(i == 'removed_list'):
                if(train[i].array[j] == 'Nao'):
                    temp.append(0)
                elif(train[i].array[j] == 'Sim'):
                    temp.append(1)
                else:
                    temp.append("nan")
            elif(i == 'priorization'):
                if(train[i].array[j] == 'Nao'):
                    temp.append(0)
                elif(train[i].array[j] == 'Sim'):
                    temp.append(1)
                else:
                    temp.append("nan")
            elif(i == 'gestation'):
                if(train[i].array[j] == 'Nao'):
                    temp.append(0)
                elif(train[i].array[j] == 'Sim'):
                    temp.append(1)
                else:
                    temp.append("nan")
            elif(i == 'prior_transplant'):
                if(train[i].array[j] == 'Nao'):
                    temp.append(0)
                elif(train[i].array[j] == 'Sim'):
                    temp.append(1)
                else:
                    temp.append("nan")
            elif(i == 'chagas'):
                if(train[i].array[j] == 'Nao'):
                    temp.append(0)
                elif(train[i].array[j] == 'Sim'):
                    temp.append(1)
                else:
                    temp.append("nan")
            elif(i == 'patient_still_on_list'):
                if(train[i].array[j] == 'Nao'):
                    temp.append(0)
                elif(train[i].array[j] == 'Sim'):
                    temp.append(1)
                else:
                    temp.append("nan")
            elif(i == 'death'):
                if(train[i].array[j] == 'Nao'):
                    temp.append(0)
                elif(train[i].array[j] == 'Sim'):
                    temp.append(1)
                else:
                    temp.append("nan")
            elif(i == 'anti.HBc'):
                if(train[i].array[j] == 'Nao'):
                    temp.append(0)
                elif(train[i].array[j] == 'Sim'):
                    temp.append(1)
                else:
                    temp.append("nan")
            elif(i == 'anti.HCV'):
                if(train[i].array[j] == 'Nao'):
                    temp.append(0)
                elif(train[i].array[j] == 'Sim'):
                    temp.append(1)
                else:
                    temp.append("nan")
            elif(i == 'agHBs'):
                if(train[i].array[j] == 'Nao'):
                    temp.append(0)
                elif(train[i].array[j] == 'Sim'):
                    temp.append(1)
                else:
                    temp.append("nan")
            else:
                temp.append(train[i].array[j])
        if(i == 'time'):
            datay.append(train[i].array[j])
    if(comp):
        datax.append(temp)
    temp = []
    comp = True


```
